# Version 1

```python
excel = openpyxl.Workbook()
# print(excel.sheetnames)
sheet = excel.active
sheet.title = 'San_Luis_Ob'
sheet.append(['FacID', 'name', 'add', 'city', 'state', 'postal', 'phone', 'date'])

options = webdriver.ChromeOptions() 
# # run browser in headless mode 
# options.headless = True 
options.add_argument('headless')
# instantiate driver 
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)

wp = []
for j in webpages:
    # webbrowser.open_new_tab(j)
    wp.append(j)
    # get the entire website content
    # driver.get(j)
    
    for k in wp:
        page = requests.get(k, verify=False)
        soup = BeautifulSoup(page.content, "html.parser")
        tr = soup.find_all('tr', class_='bodytext')

        driver.get(k)

    for i in tr:
        try:
            a = i.find('a', href=True)
            # print(a)
            name = a.text.replace('amp;  ','')
            # print(name)
            id = a['href'].split('=',2)[2]
            # print(id)
            # td = i.find_all('td')[1]
            # add = td.text.split()
            # add = ' '.join(add)
            nm = []
            nm.append(id)
            nm.append(name)
            # nm.append(add)
            
            # print(nm)

            a_tag = driver.find_element(By.LINK_TEXT, name).click()

            html_source = driver.page_source
            soup1 = BeautifulSoup(html_source, 'html.parser')
            # print(soup1)

            body = soup1.find('body')
            # print(body)

            for tag in range(7,10):
                try:
                    tr1 = body.find_all('tr')[tag]
                    tr1 = tr1.find_all('td')[1].text.replace("						", "").strip()
                    # print(tr1)
                    nm.append(tr1)
                    # print(nm)
                except:
                    pass
              
            res = []
            for l in nm:
                res.append(l.replace("\n", ""))
            # print(res)

            for s in res:
                try: 
                    res1 = res[3].split(', ')
                    res2 = res1[1].split()
                    # res.extend(res3)
                    # res.extend
                    # res3
                except:
                    pass
        
            res.insert(3, res1[0])
            res.insert(4, res2[0])
            res.insert(5, res2[1])
            res.pop(6)

            for tag in range(11,12):
                try:
                    tr1 = body.find_all('tr')[tag]
                    tr1 = tr1.find_all('td')[0]
                    tr1 = tr1.text.strip()
                    res.append(tr1)
                except:
                    pass
            
            # res = [res]  
            # col = ['ID', 'Name', 'Address', 'City', 'State', 'Postal', 'Phone', 'Date']
            # with open('C:/Users/sunpatil/Desktop/san_luis_ob5.csv', 'w', encoding='utf-8') as f:
            #     write = csv.writer(f)
            #     write.writerow(col)
            #     for row in res:
            #         write.writerows([row])
            
            sheet.append(res)
            print(res)
            driver.back()
        except:
            pass
excel.save('C:/Users/sunpatil/Desktop/san_luis_ob6.xlsx')
```
